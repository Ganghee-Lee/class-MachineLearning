```
one_hot=np.zeros((10, num_image))
for i in range(num_image) :
    one_hot[list_label[i]][i]=1
```

```
'''
Back propagation
'''
def bw_propagation() :
    global weight0
    global weight1
    global weight2
    global bias0
    global bias1
    global bias2
    d0=(a2-one_hot) @ a1.T
    c0=a2-one_hot
    b0=np.sum(c0, axis=0)

    c1=(weight2.T @ c0 * a1 * (1-a1))[1:]
    d1=c1 @ a0.T
    b1=np.sum(c1, axis=0)

    c2=(weight1.T @ c1 * a0 * (1-a0))[1:]
    d2=c2 @ input_data.T
    b2=np.sum(c2, axis=0)

    weight2-=learning_rate * d0 / input_data.shape[1]
    weight1-=learning_rate * d1 / input_data.shape[1]
    weight0-=learning_rate * d2 / input_data.shape[1]

    bias2-=learning_rate * b0 / input_data.shape[1]
    bias1-=learning_rate * b1 / input_data.shape[1]
    bias0-=learning_rate * b2 / input_data.shape[1]

def loss(output, label) :
    return np.mean(np.sum(-label*np.log(output) - (1-label)*np.log(1-output), axis=0))
def accuracy(output, label):
  return (output == label.argmax(axis = 0)).mean()
```

```
c=0
l_list=[]
acc_list=[]
for i in range(10) :
    input_data, a0, a1, a2 = fw_propagation()
    bw_propagation()
    l=loss(a2, one_hot)
    l_list.append(l)
    acc=accuracy(a2.argmax(axis=0), one_hot)
    acc_list.append(acc)
print(len(l_list))
```
