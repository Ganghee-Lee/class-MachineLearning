```
import csv
import numpy as np
import matplotlib.pyplot as plt
'''
1. Data
 - Config
 - Input Data

 2. Linear Model
 - Initialize linear model
 - Visualize data and linear model
'''

# Config
learning_rate=0.00002
opt_threshold=1e-6
t0, t1, t2, t3=0., 0., 0., 0.
theta=np.array([[t0, t1, t2, t3]])

# Load data
path = "drive/My Drive/Classroom/Machine Learning (2) 2020-1\
/class-MachineLearning/assignment04/"
train=[]
test=[]

with open(path+'data_train.csv', newline='') as myfile:
    reader  = csv.reader(myfile, delimiter=',')
    ct = 1 
    for i in reader:
        temp=[1.0, float(i[0]), float(i[1]), float(i[2]), float(i[3])]
        train.append(temp)
        #print('[', ct, ']', 'x =', i[0], ', y = ', i[1], ', z = ', i[2], ', h = ', i[3])
        ct += 1
    train_m=ct
with open(path+'data_test.csv', newline='') as myfile:
    reader  = csv.reader(myfile, delimiter=',')
    ct = 1 
    for i in reader:
        temp=[1.0, float(i[0]), float(i[1]), float(i[2]), float(i[3])]
        test.append(temp)
        #print('[', ct, ']', 'x =', i[0], ', y = ', i[1], ', z = ', i[2], ', h = ', i[3])
        ct += 1
    test_m=ct
train=np.array(train)
test=np.array(test)
train_x, train_y=train[:, :4], train[:, 4]
train_y=train_y[:,np.newaxis]

test_x, test_y=test[:,:4], test[:,4]
test_y=test_y[:,np.newaxis]

#print(train)
#print('-'*40)
#print(test)
z=train_x[:, 0]
#print(train)
#print(train[0][1]+train[0][2])
'''
plt.figure(figsize=(8, 8))
plt.scatter(z, train_y, color='black')
plt.xticks([z[0], z[10], z[-1]])
plt.yticks([train_y[0], train_y[10], train_y[-1]])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Input data')
plt.show()
'''
def train_hypothesis() : 
    return train_x.dot(theta.T)

def test_hypothesis() : 
    return test_x.dot(theta.T)
train_h=train_hypothesis()

#print(train_h)
test_h=test_hypothesis()
'''
path = "drive/My Drive/Classroom/Machine Learning (2) 2020-1\
/class-MachineLearning/assignment03/data.csv"
data = np.genfromtxt(path, delimiter=',')

x_data = data[:, 0]
y_data = data[:, 1]

plt.figure(figsize=(8, 8))
plt.scatter(x_data, y_data, color='black')
plt.show()
'''
```
