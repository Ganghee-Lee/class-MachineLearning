```
'''
3. Objective function
 - Define objective function

4. Gradient Descent
 - Define derivations
 - Update theta0 and theta1
'''

# Define objective function
def train_objective_function() :
    J1 = np.sum((h-y)**2) / (2*m)
    return J1
def test_objective_function() :
    J2 = np.sum((h-y)**2) / (2*m)
    return J2
# Define derivations
def dev() :    
    dev0=np.sum(h-y) / m
    dev1=np.sum((h-y)*x) / m
    return dev0, dev1

# Update theta0 and theta1
def gradient_descent() :
    return theta0-(learning_rate*dev0), theta1-(learning_rate*dev1)

J_list=[]
theta0_list=[]
theta1_list=[]
# Train
count=0
while(1) :
    count+=1
    J=train_objective_function()
    J_list.append(J)
    theta0_list.append(theta0)
    theta1_list.append(theta1)
    dev0, dev1 = dev()
    temp0, temp1 = theta0, theta1
    theta0, theta1 = gradient_descent()
    h=theta0+theta1*x
    if(temp0==theta0) and (temp1==theta1) :
        break
print("theta0 : ", theta0)
print("theta1 : ", theta1)
print("loss : ", J)
# Visualize data and linear model
plt.scatter(x, y, color='black')
plt.plot(x, h, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear regression results')
plt.show()
```
