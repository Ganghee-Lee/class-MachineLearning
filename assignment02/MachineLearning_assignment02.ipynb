```
import numpy as np
import matplotlib.pyplot as plt

m=1000
x=np.random.normal(0, 1, size=m)
a=20
b=-14
y_=a*x+b

n=np.random.normal(0, 3, size=m)
y=y_+n
theta0, theta1=27, 48
h=theta1*x+theta0


```

```
plt.scatter(x, y, color='black')
plt.plot(x, h, color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Input data')
plt.show()
```
