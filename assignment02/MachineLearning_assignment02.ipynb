```
import numpy as np
import matplotlib.pyplot as plt
'''
1. Data
 - Config
 - Input Data

 2. Linear Model
 - Initialize linear model
 - Visualize data and linear model
'''

# Config
m=1000
a=20
b=-14
learning_rate=0.001
theta0, theta1=27, 48

# Input data
x=np.random.normal(0, 3, size=m)
temp_y=a*x+b
n=np.random.normal(0, 3, size=m)
y=temp_y+n

# Initialize linear model
h=theta1*x+theta0

# Visualize data and linear model
plt.scatter(x, y, color='black')
plt.plot(x, h, color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Input data')
plt.show()
```
