```
import numpy as np
import matplotlib.pyplot as plt
'''
1. Data
 - Config
 - Input Data

 2. Linear Model
 - Initialize linear model
 - Visualize data and linear model
'''

# Config
m=1000
b=20
a=-14
learning_rate=0.001
theta0, theta1=-27, 48

# Input data
x=np.random.normal(0, 3, size=m)
temp_y=a*x+b
n=np.random.normal(0, 3, size=m)
y=temp_y+n

# Initialize linear model
h=theta1*x+theta0

# Visualize data and linear model
plt.scatter(x, y, color='black')
plt.plot(x, h, color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Input data')
plt.show()
```

```
'''
3. Objective function
 - Define objective function

4. Gradient Descent
 - Define derivations
 - Update theta0 and theta1
'''

# Define objective function
def objective_function() :
    J = np.sum((h-y)**2) / (2*m)
    return J

# Define derivations
def dev() :    
    dev0=np.sum(h-y) / m
    dev1=np.sum((h-y)*x) / m
    return dev0, dev1

# Update theta0 and theta1
def gradient_descent() :
    return theta0-(learning_rate*dev0), theta1-(learning_rate*dev1)

J_list=[]
theta0_list=[]
theta1_list=[]
# Train
count=0
while(1) :
    count+=1
    J=objective_function()
    J_list.append(J)
    theta0_list.append(theta0)
    theta1_list.append(theta1)
    dev0, dev1 = dev()
    temp0, temp1 = theta0, theta1
    theta0, theta1 = gradient_descent()
    h=theta0+theta1*x
    if(temp0==theta0) and (temp1==theta1) :
        break
print("theta0 : ", theta0)
print("theta1 : ", theta1)
print("loss : ", J)
# Visualize data and linear model
plt.scatter(x, y, color='black')
plt.plot(x, h, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear regression results')
plt.show()
```

```
plt.plot(J_list, color='blue')
plt.xlabel("iteration")
plt.xlim(right=100)
plt.ylabel("cost")
plt.title("Energy values")
plt.show()
```
